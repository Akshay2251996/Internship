1: Write a python program to scrape data for “Data Analyst” Job position in “Bangalore” location. You 
have to scrape the job-title, job-location, company_name, experience_required. You have to scrape first 10 
jobs data.
This task will be done in following steps:
1. First get the webpage https://www.naukri.com/
2. Enter “Data Analyst” in “Skill, Designations, Companies” field and enter “Bangalore” in “enter the 
location” field.
3. Then click the searchbutton.
4. Then scrape the data for the first 10 jobs results you get.
5. Finally create a dataframe of the scraped data

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
from selenium.webdriver.common.by import By
import time
```

```python
# Connect to the driver
driver=webdriver.Chrome(r"C:\Users\user\Desktop\internship\webdriver\chromedriver.exe")
```

```python
# opening naukri page on automated chrome browser
driver.get("https://www.naukri.com/")
```

```python
# entering designation and skill
designation=driver.find_element(By.CLASS_NAME,"suggestor-input ")
designation.send_keys('Data Analyst')
```

```python
# entering location
location=driver.find_element(By.XPATH,"/html/body/div[1]/div[6]/div/div/div[5]/div/div/div/input")
location.send_keys('Bangalore')
```

```python
# searching
serach=driver.find_element(By.CLASS_NAME,"qsbSubmit")
serach.click()
```

```python
job_title=[]
job_location=[]
company_name=[]
experience_required=[]
```

```python
#scrapping job title
job_tags=driver.find_elements(By.XPATH,'//a[@class="title ellipsis"]')
for i in job_tags[0:10]:
    title=i.text
    job_title.append(title)
    

# scrapping job location
location_tags=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft locWdth"]')
for i in location_tags[0:10]:
    loc=i.text
    job_location.append(loc)
    
# scrapping company name
company_tags=driver.find_elements(By.XPATH,'//a[@class="subTitle ellipsis fleft"]')
for i in company_tags[0:10]:
    name=i.text
    company_name.append(name)
    
# scrapping experience required
experience_tags=driver.find_elements(By.XPATH,'//span[@class="ellipsis fleft expwdth"]')
for i in experience_tags[0:10]:
    exp=i.text
    experience_required.append(exp)
```

```python
# checking for length
print(len(job_title),len(job_location),len(company_name),len(experience_required))
```

```python
#for Data frame
import pandas as pd
df=pd.DataFrame({'Title':job_title,'Location':job_location,'Company_name':company_name,'Experience':experience_required})
df
```

Note: All of the above steps have to be done in code. No step is to be done manually.

Q8: Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:
1. First get the webpagehttps://www.azquotes.com/
2. Click on TopQuotes
3. Than scrap a) Quote b) Author c) Type Of Quotes

```python
# Connect to the driver
driver=webdriver.Chrome(r"C:\Users\user\Desktop\internship\webdriver\chromedriver.exe")
```

```python
# opening quotes page on automated chrome browser
driver.get("https://www.azquotes.com/")
```

```python
# top quotes
top=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
top.click()
```

```python
quote=[]
author=[]
qt_type=[]
```

```python
#scrapping for quote
quote_tags=driver.find_elements(By.XPATH,'//a[@class="title"]')
for i in quote_tags:
    qt=i.text
    quote.append(qt)
    

# scrapping author
author_tags=driver.find_elements(By.XPATH,'//div[@class="author"]')
for i in author_tags:
    at=i.text
    author.append(at)
    
# scrapping quote type
type_tags=driver.find_elements(By.XPATH,'//div[@class="tags"]')
for i in type_tags:
    tp=i.text
    qt_type.append(tp)
```

```python
print(len(quote),len(author),len(qt_type))
```

```python
start=0
end=10
for page in range(start,end):
    quotes=driver.find_elements(By.XPATH,'//a[@class="title"]')
    for i in quotes:
        quote.append(i.text)
        
    authors=driver.find_elements(By.XPATH,'//div[@class="author"]')
    for i in authors:
        author.append(i.text)
        
    types=driver.find_elements(By.XPATH,'//div[@class="tags"]')
    for i in types:
        qt_type.append(i.text)
    next_button=driver.find_elements(By.XPATH,'//li[@class="next"]')
    next_button
    time.sleep(3)
```

```python
print(len(quote),len(author),len(qt_type))
```

```python
df=pd.DataFrame({'Quotes':quote[0:1000],'Authors':author[0:1000],'Type Of Quotes':qt_type[0:1000]})
df
```
